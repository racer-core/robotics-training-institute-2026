### What is Dead Reckoning?
**Dead reckoning** is a way of navigating by keeping track of the moves you make, without using any outside reference like GPS or landmarks. If you know your starting point, and you carefully track every direction and distance you travel, you can figure out where you end up.

Sailors used dead reckoning for centuries: starting from a known port, they recorded their heading and speed over time to estimate their position at sea.  
*Your robot has no GPS either — so this is exactly how it will navigate today.*

### How does *our robot* do dead reckoning?
You already have all the pieces from the last two notebooks:

 + **Heading** (from A2): `drive_robot()` lets you point the robot in an exact direction — 0, 90, 180, 270 degrees, or anything in between.
 + **Distance** (from A3): your calibration model `time_for_distance()` converts a distance in meters into a drive time.

Put them together and you can command the robot: *"face this heading, and drive far enough to cover this distance."*    
Chain several of these together and you have a **route** — a sequence of legs that carries the robot from a start point to a goal, one straight segment at a time.

**Real Robotics Engineering:** Dead reckoning is *open-loop* navigation — the robot follows your commands blindly without checking whether it actually arrived.  
Small errors in each leg (a slightly-off heading, a wheel slipping on carpet) **add up** over the whole route. Real robots fight this by adding sensors (cameras, encoders, GPS) that *correct* the estimate as they go. You'll feel this error accumulation today — that's the whole point of the challenge.

### A quick reminder about heading
When you call `drive_robot()`, the **heading** is measured from the direction the robot was facing when it started. Picture a compass drawn on the floor under the robot:

 + **90** = to the right
 + **180** = straight back
 + **270** = to the left
